```python
import openpyxl
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df_country.shape
```

```python
df.columns
```

```python
final_df = pd.merge(df,df_country,on='Country Code', how = 'left')
```

```python
final_df.head(2)
```

```python
final_df.columns
```

```python
country_names = final_df.Country.value_counts().index
country_names
```

```python
country_values = final_df.Country.value_counts().values
country_values
```

```python
plt.pie(x = country_values[:3] ,labels= country_names[:3],autopct = '%1.2f')
```

Observation : Zomato max transcations are from india

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns = {0:'Rating count'})
ratings
```

## Observation

1. maximum number people have not rated any restaurant.
2. Average nad good rating are the second most rated restaurant.

```python
plt.rcParams['figure.figsize'] = (12,6)
sns.barplot(x = 'Aggregate rating', y = 'Rating count',data = ratings)
```

```python
plt.rcParams['figure.figsize'] = (12,6)
sns.barplot(x = 'Aggregate rating', y = 'Rating count',data = ratings,hue = 'Rating color', palette = ['grey','red','orange','yellow','green','green'])
```

```python
sns.countplot(x = 'Rating color',data = ratings,palette = ['grey','red','orange','yellow','green','green'] )
```
